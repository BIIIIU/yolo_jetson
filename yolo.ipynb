```python
import pandas as pd

# 请替换为您的CSV文件路径
csv_file_path = "/home/niqi/workspace/YOLO/pulserate.csv"

# 从CSV文件读取数据
chunksize = 1

# 创建一个可迭代的对象，每次迭代返回指定数量的行
csv_reader = pd.read_csv(csv_file_path, header=None, chunksize=chunksize, skiprows=1)
all_rows = []
# # 逐行处理数据
for chunk in csv_reader:
    all_rows.extend(chunk.values.tolist())
    # 在这里进行处理，chunk 是一个包含一行数据的 DataFrame
    # print(chunk.shape)
num_samples = len(all_rows)
# for row in all_rows:
#     print(row)
```
